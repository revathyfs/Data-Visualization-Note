
# ðŸ“Š Data Visualization in Data Science

## 1. **Definition**

Data visualization is the process of representing data and information in a **graphical or pictorial format**. Instead of analyzing rows and columns of raw data, visualization uses charts, graphs, maps, and dashboards to make patterns, trends, and insights easier to understand.

---

## 2. **Why Visualization is Important?**

* **Simplifies complex data** â†’ Easier to interpret than tables.
* **Reveals hidden patterns** â†’ Helps detect trends, correlations, and outliers.
* **Improves communication** â†’ Explains insights to non-technical stakeholders.
* **Supports decision-making** â†’ Quick, data-driven decisions with visual clarity.

---

## 3. **Types of Data Visualization**

### ðŸ”¹ a) **Basic Charts**

* **Bar Chart** â€“ Compare categories (e.g., sales by product).
* **Pie Chart** â€“ Show proportions of a whole.
* **Histogram** â€“ Distribution of continuous variables.
* **Line Chart** â€“ Trends over time.

### ðŸ”¹ b) **Advanced Visuals**

* **Heatmaps** â€“ Show correlation or intensity.
* **Boxplots** â€“ Show distribution, outliers.
* **Scatter Plots** â€“ Show relationships between variables.
* **Tree Maps** â€“ Show hierarchical data.

### ðŸ”¹ c) **Interactive Visualization**

* Dashboards (Power BI, Tableau, Plotly Dash, Streamlit).
* Interactive charts where users can filter and drill down.

---

## 4. **Tools & Libraries**

### ðŸ“Œ Programming

* **Python**: `matplotlib`, `seaborn`, `plotly`, `bokeh`
* **R**: `ggplot2`, `lattice`

### ðŸ“Œ Business Intelligence

* Tableau, Power BI, QlikView

---

## 5. **Steps in Data Visualization Workflow**

1. **Understand the Data** â€“ Numerical? Categorical? Mixed?
2. **Choose the Right Chart** â€“ Based on what you want to highlight.
3. **Clean & Preprocess Data** â€“ Handle missing values, scaling.
4. **Build Visualization** â€“ Using libraries/tools.
5. **Interpret & Communicate** â€“ Tell the story behind the chart.

---

## 6. **Best Practices**

* **Keep it simple** â†’ Avoid clutter and 3D effects.
* **Use correct chart types** â†’ E.g., donâ€™t use pie chart for too many categories.
* **Label clearly** â†’ Titles, legends, axis labels.
* **Highlight key insights** â†’ Use color, annotations.

---

## 7. **Real-World Applications**

* **Business** â€“ Sales dashboards, customer segmentation.
* **Healthcare** â€“ Patient statistics, outbreak visualization.
* **Finance** â€“ Stock market trends, risk analysis.
* **Government** â€“ Population distribution, election results.
* **Sports** â€“ Player performance, match analysis.

---

## 1. **What is Matplotlib?**

* **Matplotlib** is a Python **plotting library** used for creating static, animated, and interactive visualizations.
* It provides **control at a low level** (like plotting in MATLAB), making it very flexible.
* It is often used together with **NumPy** and **Pandas** for handling data.

## 4. **Common Plot Types**

### 4.1. **Line Chart**

```python
plt.plot(x, y, color=..., linestyle=..., marker=..., label=...)
```

### 4.2. **Bar Chart**

```python

```

### 4.3. **Horizontal Bar Chart**

```python

```

### 4.4. **Histogram**

```python
plt.hist(data, bins=..., color=..., edgecolor=...)
```

### 4.5. **Scatter Plot**

```python
plt.scatter(x, y, color=..., marker=..., s=...)
```

### 4.6. **Pie Chart**

```python
plt.pie(sizes, labels=..., autopct=..., startangle=..., colors=...)
```

### 4.7. **Stacked Bar Chart**

```python
plt.bar(x, A, label=...)
plt.bar(x, B, bottom=A, label=...)
```

### 4.8. **Box Plot**

```python

```

### 4.9. **Multiple Plots (Subplots)**

```python
plt.subplot(rows, cols, index)
```

### 4.10. **Area Chart**

```python

plt.plot(x, y, color=...)
```

### 4.11. **Stacked Area Chart**

```python

```

````python
import matplotlib.pyplot as plt
import numpy as np

#plot1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)

#plot2:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
````

What happens step by step:

1. `plt.subplot(1,2,1)` â†’ divides the figure into 1 row Ã— 2 columns and activates the **first subplot**.

   * You plot `x=[0,1,2,3], y=[3,8,1,10]` here.

2. `plt.subplot(1,2,2)` â†’ activates the **second subplot** in the same row.

   * You plot `x=[0,1,2,3], y=[10,20,30,40]` here.

3. `plt.show()` â†’ displays both plots in one window.

---

âœ… Expected output:

* **Left plot (first subplot):** line graph going through (0,3), (1,8), (2,1), (3,10).
* **Right plot (second subplot):** line graph going through (0,10), (1,20), (2,30), (3,40).
