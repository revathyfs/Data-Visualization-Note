```python
import matplotlib.pyplot as plt
import numpy as np
#if using a jupyter notbook
#%matplotlib inline

x=np.arange(0,6*np.pi,0.1)  #start,stop,step
y=np.sin(x)

plt.plot(x,y)
plt.show()

x=np.arange(0,4*np.pi,0.1)  #start,stop,step
y=np.sin(x)
z=np.cos(x)

plt.plot(x,y,x,z)
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1) #start,stop,step
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,x,z)
plt.xlabel('x values from 0 to 4pi') #string must be enclosed with quotes ''
plt.ylabel('sin(x)and cos(x)')
plt.title('plot of sin and cos from 0 to 4pi')
plt.legend(['sin(x)','cos(x)']) #legend entries as seperate strings in a list
plt.show()
```

# Bar Plot

```python
# Bar Plot
import matplotlib.pyplot as plt
import numpy as np

grades=["A","B","C","D"]
test1=[3,10,40,35]
#test2=[40,50,30,25]
#test3=[35,55,60,75]

plt.bar(grades,test1)
plt.show()
```

## Giving title and label

```python
grades=["A","B","C","D"]
test1=[3,10,40,35]
#test2=[40,50,30,25]
#test3=[35,55,60,75]

plt.bar(grades,test1)
plt.title('BAR PLOT')
plt.xlabel('Grades')
plt.ylabel('No of students')
plt.show()
```

```python
#giving fontcolour,fontsizes
grades=["A","B","C","D"]
test1=[3,10,40,35]
#test2=[40,50,30,25]
#test3=[35,55,60,75]
plt.bar(grades,test1)
plt.title('BAR PLOT',fontsize=15,color='firebrick')

plt.xlabel('Grades',fontsize=15,color='magenta')
plt.ylabel("No: of students",fontsize=15,color='red')
plt.show()
```

## Giving bar width and alignment

```python
# giving bar width and alignment

grades=["A","B","C","D"]
test1=[3,10,40,35]
#test2=[40,50,30,25]
#test3=[35,55,60,75]
plt.bar(grades,test1,color='lime',edgecolor='red',width=0.5,align="center")
plt.title('BAR PLOT',fontsize=15,color='firebrick')

plt.xlabel('Grades',fontsize=15,color='magenta')
plt.ylabel("No: of students",fontsize=15,color='red')
plt.show()
```

## Line width and line style 

```python
# linewidth and linestyle

grades=["A","B","C","D"]
test1=[3,10,40,35]
#test2=[40,50,30,25]
#test3=[35,55,60,75]
plt.bar(grades,test1,color='lime',edgecolor='red',width=0.5,align="center",linestyle="--",linewidth=6)
plt.title('BAR PLOT',fontsize=15,color='firebrick')

plt.xlabel('Grades',fontsize=15,color='magenta')
plt.ylabel("No: of students",fontsize=15,color='red')
plt.show()
```

## Three Bar

```python
import numpy as np
import matplotlib.pyplot as plt
grades=["A","B","C","D"]
test1=[3,10,40,35]
test2=[40,50,30,25]
test3=[35,55,60,75]

index=np.arange(len(grades))

plt.bar(index,test1,width=0.2,align='center',color='red',edgecolor="blue",linewidth=2,label='test1')
plt.bar(index+0.2,test2,width=0.2,align='center',color='lime',edgecolor="k",linewidth=2,label='test2')
plt.bar(index-0.2,test3,width=0.2,align='center',color='cyan',edgecolor="m",linewidth=2,label='test3')
plt.title('BAR PLOT',fontsize=15,color='firebrick')

plt.xlabel('Grades',fontsize=15,color='magenta')
plt.ylabel("No: of students",fontsize=15,color='red')
plt.legend()
plt.show()
print(len(grades))
print(index)
```

## Background set

```python
import numpy as np
import matplotlib.pyplot as plt

grades=["A","B","C","D"]
test1=[3,10,40,35]
test2=[40,50,30,25]
test3=[35,55,60,75]
#for background
plt.figure(figsize=(10,10),facecolor='lightblue',edgecolor='pink',linewidth=5)
plt.style.use("Solarize_Light2")
index=np.arange(len(grades))

plt.bar(index,test1,width=0.2,align='center',color='red',edgecolor='k',linewidth=2,label='test1')
plt.bar(index+0.2,test2,width=0.2,align='center',color='lime',edgecolor='k',linewidth=2,label='test2')
plt.bar(index-0.2,test3,width=0.2,align='center',color='cyan',edgecolor='k',linewidth=2,label='test3')

plt.title('BAR PLOT',fontsize=15,color='firebrick')

plt.xlabel('Grades',fontsize=15,color='magenta')
plt.ylabel("No: of students",fontsize=15,color='red')
plt.legend()
plt.show()
```
