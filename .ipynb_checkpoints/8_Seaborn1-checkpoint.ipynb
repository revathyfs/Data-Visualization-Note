
# ðŸŽ¨ Seaborn

## 1. **Definition**

* **Seaborn** is a Python data visualization library built on top of **Matplotlib**.
* It provides **high-level, beautiful, and easy-to-use** interfaces for creating statistical graphics.
* Unlike Matplotlib, Seaborn comes with **better defaults**, works directly with **Pandas DataFrames**, and has **built-in statistical plots**.

---

## 2. **Why Seaborn?**

* Cleaner and more attractive plots than Matplotlib.
* Works directly with **DataFrame columns** (no need to pass lists manually).
* Provides advanced plots (heatmaps, violin plots, pair plots).
* Best for **statistical data visualization**.

---

## 3. **Seaborn Chart Types & Syntax**

### ðŸ”¹ Line Plot

```python
sns.lineplot(x=..., y=..., data=..., hue=..., style=..., markers=...)
```

### ðŸ”¹ Bar Chart

```python
sns.barplot(x=..., y=..., data=..., hue=..., ci=...)
```

### ðŸ”¹ Count Plot (for categorical frequency)

```python
sns.countplot(x=..., data=..., hue=...)
```

### ðŸ”¹ Histogram

```python
sns.histplot(data=..., x=..., bins=..., kde=..., hue=...)
```

### ðŸ”¹ KDE Plot (Density)

```python
sns.kdeplot(data=..., x=..., shade=..., multiple=...)
```

### ðŸ”¹ Scatter Plot

```python
sns.scatterplot(x=..., y=..., data=..., hue=..., style=..., size=...)
```

### ðŸ”¹ Regression Plot (Line + Scatter)

```python
sns.regplot(x=..., y=..., data=..., ci=..., scatter=...)
```

### ðŸ”¹ Box Plot

```python
sns.boxplot(x=..., y=..., data=..., hue=...)
```

### ðŸ”¹ Violin Plot

```python
sns.violinplot(x=..., y=..., data=..., hue=..., split=...)
```

### ðŸ”¹ Swarm Plot (points + distribution)

```python
sns.swarmplot(x=..., y=..., data=..., hue=...)
```

### ðŸ”¹ Strip Plot (jittered scatter for categorical)

```python
sns.stripplot(x=..., y=..., data=..., jitter=..., hue=...)
```

### ðŸ”¹ Pair Plot (relationships between multiple features)

```python
sns.pairplot(data=..., hue=..., kind=...)
```

### ðŸ”¹ Heatmap (matrix visualization)

```python
sns.heatmap(data=..., annot=..., cmap=..., linewidths=...)
```

### ðŸ”¹ Clustermap (heatmap with clustering)

```python
sns.clustermap(data=..., cmap=..., standard_scale=...)
```

---

ðŸ“Œ **Customization (common across all plots)**

```python
plt.title("...")
plt.xlabel("...")
plt.ylabel("...")
sns.set_style("whitegrid")     # or "darkgrid", "white", "dark"
```

# ðŸŽ¨ When to Use Seaborn?

Seaborn is best when you want **statistical, high-level, and attractive visualizations** with minimal effort.

---

## ðŸ”¹ 1. **When you want beautiful plots quickly**

* Seaborn has **better default styles** than Matplotlib.
* Plots are clean, modern, and publication-ready without much customization.

ðŸ‘‰ Example use: Presenting a project report or dashboard.

---

## ðŸ”¹ 2. **When working with Pandas DataFrames**

* Seaborn accepts DataFrame **directly** (`data=...`).
* You can use column names instead of passing separate arrays.

ðŸ‘‰ Example use: `sns.barplot(x="species", y="sepal_length", data=iris)`

---

## ðŸ”¹ 3. **When you need statistical visualization**

* Seaborn automatically computes and displays:

  * **Confidence intervals** (in `barplot`, `lineplot`)
  * **Distribution** (`histplot`, `kdeplot`)
  * **Relationships** (`regplot`, `pairplot`)

ðŸ‘‰ Example use: Checking correlation between features in Titanic dataset.

---

## ðŸ”¹ 4. **When you need advanced plots (not in Matplotlib directly)**

* **Heatmaps** (correlation, confusion matrix)
* **Violin plots** (distribution + boxplot hybrid)
* **Pair plots** (relationships between multiple variables)
* **Clustermaps** (heatmap + hierarchical clustering)

ðŸ‘‰ Example use: Exploring datasets before applying ML models.

---

## ðŸ”¹ 5. **When exploring data (EDA â€“ Exploratory Data Analysis)**

* Seaborn is designed for **EDA** because it helps:

  * Spot trends, outliers, and distributions.
  * Understand categorical vs numerical relationships.
  * Compare multiple groups easily.

ðŸ‘‰ Example use: Analyzing Titanic dataset survival by gender/class.

---

## ðŸ”¹ 6. **When you want built-in themes**

```python
sns.set_style("whitegrid")   # or "darkgrid", "white", "ticks"
```

* Styles make charts look polished instantly.
* Saves time compared to Matplotlib customizations.

---

# âœ… Summary

ðŸ‘‰ **Use Seaborn when:**

* You need **statistical plots** (distributions, regression, heatmaps).
* You are working with **DataFrames**.
* You want **beautiful plots with minimal code**.
* Youâ€™re doing **EDA or presenting results**.

ðŸ‘‰ **Use Matplotlib when:**

* You need **low-level control** and customization.
* You want to build **custom/complex** plots.
* Youâ€™re working in **scientific/engineering** contexts where every detail matters.

````python
# we can plot a single by two different graphs in seaborn
# jointplot() 
# It is a Seaborn function that creates a combined plot showing the relationship between two variables,along with their individual distributions.
# It's a great way to:
# Visualize how two variables are related (e.g. correlation).
# Understand the distribution (spread, skew, outliers) of each variable.
# sns.jointplot(data=your_dataframe, x='column1', y='column2')
# We get:
# A main plot in the center â€” usually a scatter plot by default.
# Marginal plots â€” histograms or KDEs â€” on the top (for x) and right (for y).

# parameter     ->   Description

# data          ->   Your DataFrame (e.g., df)
# x, y	        ->   Column names to plot
# kind	        ->   Type of central plot ('scatter', 'reg', 'hex', 'kde', etc.)
# height        ->   Size of the figure
# hue           ->   Grouping variable (available in recent versions of Seaborn)
# marginal_kws  ->   Custom settings for the marginal plots (e.g., add kde=True)

# It's exploratory â€” you can quickly see patterns, clusters, or correlations.
# Helps identify outliers or non-linear relationships.
# Adds context by showing both the relationship and distribution in one figure.

sns.jointplot(data=df,x='total_bill',y='tip')
````

````python
# to compare numerical columns we use pairplot()
# sns.pairplot(dataset_variable_name)
# It compares each numerical columns in the table
# In Seaborn it is a powerful tool used to visualize pairwise relationships between all numerical columns in a DataFrame.
# It helps you:
# See correlations between numerical variables
# Spot trends, clusters, and outliers
# Quickly understand how different variables relate to each other
# Creates a grid of scatter plots for every pair of numerical columns
# Places histograms or KDEs on the diagonal (to show the distribution of each variable)
# For Example:
# If your DataFrame df has 4 numerical columns: A, B, C, D
# pairplot(df) creates a 4Ã—4 grid:
# Each cell shows a scatter plot of one variable vs another (e.g., A vs B)
# Diagonal cells show the distribution of a single variable (e.g., histogram of A)

# Parameter	              ->  Description
# hue='column'	          ->  Color points by a categorical variable
# kind='scatter'/'kde'	  ->  Type of plots (scatter is default)
# diag_kind='hist'/'kde'  ->  Controls the diagonal plot type
# markers	              ->  Custom markers for different hue levels
# palette	              ->  Set color scheme

# Use it during exploratory data analysis (EDA) to:
# Detect patterns or relationships
# Check for linear or non-linear associations
# Compare distributions across categories (with hue)

sns.pairplot(df)
````

````python
#countplot()
# used to get valueplot
# it is used to create a bar plot showing the count (frequency) of each unique value in a categorical column.
# It takes a categorical variable
# Counts how many times each category appears
# Plots each category as a bar, where the height = count
# Use countplot():
# To understand the distribution of categories in a dataset
# To quickly check for imbalanced classes
# To visualize group sizes (like number of male vs female customers)

# Parameter	-> Purpose
# x or y	-> Axis that holds the categorical variable
# data	    -> The DataFrame you're using
# hue	    -> Splits bars based on a second categorical variable
# order	    -> Custom order of categories
# palette	-> Set color scheme
# orient	-> Plot bars horizontally ('h') instead of vertically

# sns.countplot(x='column_name', data=your_dataframe)

sns.countplot(x='sex',data=df)
````

````python
sns.countplot(x='sex',data=df)

# for style the plot
# set_style() is a Seaborn function used to change the background style of your plots.
# It controls the appearance of things like: Gridlines,Axis lines,Background color
# It's purely for visual styling â€” it doesn't affect the data or chart type.

# sns.set_style('style_name')

# Style Name  ->	Description
# 'darkgrid'  ->	Dark background with grid lines (default) â€” great for data visibility
# 'whitegrid' ->	White background with grid lines â€” useful for plots with numeric data
# 'dark'	  ->    Dark background without grid lines â€” clean and simple
# 'white'     ->	White background without grid lines â€” very minimal
# 'ticks'     ->	White background with ticks on axes â€” professional and precise look

# Use It:
# To improve readability and match presentation style
# Different plots benefit from different styles (e.g., whitegrid is great for bar plots)
# Helps make your plots look more polished and professional

sns.set_style('darkgrid')
````

````python
sns.countplot(x='time',data=df,hue='smoker')

# hue Splits each bar in the chart based on the 'smoker' column (e.g., "Yes" vs "No"), using different colors for each category.
# In Seaborn, the hue parameter is used to add a third variable to your plot by splitting data within categories using color.

# When you set hue='column_name', Seaborn:
# Groups the data based on that column.
# Assigns different colors to each group.
# Draws sub-bars, points, or lines within the main category.
# This makes your plot more informative, showing how another variable changes across the main categories.

# Use hue when:
# You want to compare subgroups within categories.
# You need to add more detail without cluttering the plot.
# Youâ€™re doing exploratory data analysis and want deeper insights.

# The main axis (x or y) shows one variable.
# hue shows a second category, using color to compare groups within each bar or point.

# Works With Many Plot Types:
# countplot() â€“ bar charts
# scatterplot() â€“ colored dots
# boxplot(), violinplot() â€“ grouped by color
# histplot() â€“ layered histograms
````

````python
# barplot()

# barplot() in Seaborn is used to show the average (or another statistic) of a numerical variable for each category of a categorical variable.
# Unlike countplot() (which just counts occurrences), barplot() actually computes a summary statistic â€” by default, the mean.
# Groups the data by categories (like smoker or day)
# Calculates a statistic (by default, the mean) for each group
# Draws bars to represent those average values

# Parameter  -> What it does
# x / y	     -> The variables being plotted (numerical and categorical)
# data	     -> The DataFrame source
# hue	     -> Adds a second category for comparison using color
# estimator	 -> Chooses the summary function (mean, median, etc.)

# Use barplot():
# Compare averages between categories
# Add a second layer of comparison using color (hue)
# Show confidence intervals (optional)

# sns.barplot(x='numerical_column', y='categorical_column', data=df, hue='another_categorical_column')
    
sns.barplot(x='total_bill',y='smoker',data=df,hue='sex')
````

````python
# boxplot()

# A boxplot (also called a box-and-whisker plot) is a type of chart that shows the distribution of a numerical variable and helps you:
# See how values are spread
# Find the median
# Detect outliers
# Itâ€™s commonly used in exploratory data analysis (EDA).

# A boxplot displays five key summary statistics:
# Minimum           â€“  the smallest value (excluding outliers)
# 1st Quartile (Q1) â€“  25% of the data falls below this value
# Median (Q2)       â€“  the middle value of the data
# 3rd Quartile (Q3) â€“  75% of the data falls below this value
# Maximum           â€“  the largest value (excluding outliers)

# Outliers (values that are unusually high or low) are shown as dots.

# Use a boxplot to:
# Compare values across groups (like days, genders, or categories)
# Spot outliers quickly
# Understand the spread and center of your data

# Feature       ->	Purpose
# Box	        ->  Shows middle 50% of the data (Q1 to Q3)
# Line in box	->  Median (middle value)
# Whiskers	    ->  Range of most data (excluding outliers)
# Dots	        ->  Outliers
# used to dedect outlayer

sns.boxplot(x='day',y='total_bill',data=df)
````

````python
# violinplot()
# A violin plot is a type of data visualization that shows the distribution of a numerical variable across different categories.
# It is similar to a boxplot, but it also includes a density curve, which makes it easier to see:
# The shape of the data distribution
# Whether the data is skewed
# If there are multiple peaks

# A violin plot combines:
# A boxplot: showing the median and quartiles
# A KDE (Kernel Density Estimate) plot: showing the dataâ€™s distribution as a smooth curve
# This helps you understand:
# Where the data is concentrated
# How spread out it is
# If it has any unusual patterns or outliers

# We can visually compare:
# The average length
# The spread of values
# Peaks or clusters in the data

# Use a violin plot when you want to:
# See more than just the average â€” like how values are distributed
# Compare distributions across categories
# Spot outliers, skewness, or multiple peaks

# Feature	        ->Shows
# Shape (violin)	->Data distribution/density
# Width             ->Where values are more common
# Middle line	    ->Median
# Box (inside)	    ->Quartiles (optional)
# Dots (optional)	->Raw data points or outliers

# sns.violinplot(x='category_column', y='numeric_column', data=your_dataframe)

sns.violinplot(x='species',y='sepal_length',data=d1)
````

````python
# stripplot()

# A strip plot is a simple type of plot that shows individual data points for a numerical variable across different categories.
# Think of it as a scatter plot where one axis (usually the x-axis) is categorical and the other is numeric.

# Each dot in the plot represents one data point.
# It's used to:
# Show exact values (not just summaries like averages or medians)
# See clustering of data
# Visualize how values are spread across categories

# A strip plot is a simple plot that shows individual data points along a category axis.
# Itâ€™s useful for:
# Seeing exact values of observations
# Understanding distribution and clustering
# Spotting overlapping data points

# The function sns.stripplot() creates a scatter plot where:
# x-axis (or y-axis) is a categorical variable
# y-axis (or x-axis) is a continuous/numerical variable
# Each point represents a single data value
# Points are jittered (slightly spread out) to avoid overlapping


# sns.stripplot(x='category', y='value', data=your_dataframe)

# If many points have the same or very similar values, they overlap.
# jitter=True spreads the points slightly left and right (or up/down) to avoid overlapping, so you can see the density of data better.
# jitter is used to avoid overlapping of datapoints (shows distance between datapoints)

# Use it when you want to:
# See every individual data point
# Avoid summarizing data too much (like with barplots or boxplots)
# Compare distributions visually
# Show a third variable using color (with hue)

# Feature           ->	Purpose
# One dot	        ->  One raw data point
# Categorical x/y	->  Groups the data
# Jitter	        ->  Prevents overlap so points are visible
# Hue	            ->  Adds color for a third variable
# Good for          ->  Raw data visualization and small-to-medium datasets
    
sns.stripplot(x='species',y='sepal_length',data=d1,jitter=True,hue='petal_width')
````

````python
# histplot()

# A histogram is a type of graph that shows how often different ranges of values occur in a dataset.
# It divides numeric data into equal-sized intervals (called bins) and counts how many values fall into each one.

# In Seaborn, histplot() is the function used to create histograms.
# Itâ€™s used to visualize the distribution of a single numerical variable.
# Groups continuous data into intervals (bins)
# Shows how frequently values occur in each interval
# Optionally overlays a KDE (Kernel Density Estimate) curve to show the dataâ€™s shape more smoothly

# The x-axis shows the range of values (e.g., test scores, prices).
# The y-axis shows how many data points fall into each bin (frequency).
# Optionally, you can add a KDE curve â€” a smooth line that estimates the shape of the distribution.

# Parameter	->  Description
# data	    ->  Your numerical data
# bins	    ->  Number of intervals (optional)
# kde=True	->  Adds a KDE (smooth curve over histogram)(Kernel Density Estimate)
# color	    ->  Sets color of bars
# stat	    ->  'count', 'density', or 'probability' (default is 'count')

# Use of a Histogram
# To understand the distribution of data (e.g., is it normal, skewed, etc.)
# To spot concentrations and gaps in values
# To see if the data has outliers or multiple peaks


# Purpose    : Show how values are distributed
# Best for   : Single numerical variables
# Helpful for: Detecting patterns, skewness, and outliers
    
import numpy as np
sns.set(style='white')
rs=np.random.RandomState(10)
d=rs.normal(size=100)
sns.histplot(d,kde=True,color='g')
````

````python
# lineplot()

# sns.lineplot() is a function in Seaborn used to create line charts, which are ideal 
# for visualizing trends over time or relationships between two continuous variables.

# A line plot connects data points with lines, usually to:
# Show how a variable changes over another (often time)
# Highlight trends, patterns, or relationships

# sns.lineplot(x='x_column', y='y_column', data=your_dataframe)

# x: Variable for the horizontal axis (e.g., time, categories)
# y: Variable for the vertical axis (e.g., sales, temperature)
# data: Your DataFrame

# Parameter	   ->  Purpose
# hue	       ->  Adds color lines for sub-groups (e.g., gender, region)
# style	       ->  Different line styles for categories
# markers=True ->  Shows individual data points along the line
# palette	   ->  Changes color theme

# Use it when:
# You want to track changes or trends across time or values
# You're comparing multiple groups with different lines
# You want a clear view of continuity in the data

# Feature	        ->  Purpose
# Line plot	        ->  Shows trends and changes
# Works best for	->  Time series or continuous data
# Can include	    ->  Categories with hue, style, markers

sns.lineplot(x='species',y='petal_length',hue='petal_width',data=d1)
````

````python
sns.lineplot(x='species',y='petal_length',hue='petal_width',data=d1,style='sepal_length') #add style
````

````python
# lmplot()

# lmplot() stands for Linear Model Plot.
# It's a Seaborn function used to visualize linear relationships between two variables, 
# along with the best-fit regression line.
# It follows a linear regression model

# Plots a scatterplot of the data points
# Adds a regression line that shows the best-fit linear relationship
# Optionally shows a confidence interval (usually 95%) around the line

# sns.lmplot(x='x_column', y='y_column', data=your_dataframe)

# x: Independent variable (predictor)
# y: Dependent variable (outcome)
# data: The dataset (DataFrame)

# Parameter	-> Description
# hue	    -> Adds separate regression lines by category (color-coded)
# col/row	-> Creates multiple subplots (facets)
# markers	-> Customize scatterplot marker style
# ci	    -> Set the size of the confidence interval (e.g., ci=None to hide it)
# fit_reg	-> Show/hide the regression line (fit_reg=False hides the line)

# Use lmplot() when you want to:
# Explore relationships between two continuous variables
# Visualize trends or possible correlations
# Compare regression lines across different groups (using hue, col, or row)
# mostly used in linear regression.

# Feature	            -> Purpose
# Scatter + Regression	-> Shows trend and individual data points
# Confidence interval	-> Indicates uncertainty in the line
# Hue/Col/Row	        -> Enables subgroup comparisons
# Great for	            -> Exploratory data analysis & model diagnostics

sns.lmplot(x='sepal_length',y='petal_length',data=d1)
````

````python
# distplot() was part of Seaborn, a statistical data visualization library built on top of Matplotlib. However, as of Seaborn version 0.11.0, 
# sns.distplot() has been deprecated in favor of newer and more explicit functions.

# This function used to plot univariate distributions with optional kernel density estimation (KDE).
# Use sns.histplot() and/or sns.kdeplot() depending on your need

sns.distplot(d1['petal_length'], kde=True, bins=20, color='skyblue')
````

````python
sns.distplot(d1['petal_length'], hist=True, bins=20, color='skyblue')
````

````python
sns.distplot(d1['petal_length'], hist=False, bins=20, color='skyblue')
````

````python
# swarmplot() is a function in Seaborn used to create a categorical scatter plot â€” also known as a swarm plot. 
# It shows each observation as a point, arranged to avoid overlap (unlike strip plots, which can overlap heavily). 
# This gives you a clear view of the distribution and density of your data.

# sns.swarmplot(x='x_column', y='y_column', data=your_dataframe)

# Parameters:
# x, y: Names of variables in data or arrays.
# data: DataFrame or structured data.
# hue: Optional categorical variable for color separation.
# palette: Color palette.
# size: Marker size.
# dodge: If hue is set, whether to separate the points horizontally.

# Swarm plots can be slow with large datasets.
# For large datasets, consider sns.stripplot() with jitter=True or violinplot()/boxplot() instead.


sns.swarmplot(x='species', y='petal_length', data=d1)
````
