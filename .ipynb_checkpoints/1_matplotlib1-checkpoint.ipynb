```python
import matplotlib.pyplot as plt
import numpy as np
x=[2,4,6,8]
y=[3,6,9,10]
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
xpoints=np.array([0,6])
ypoints=np.array([0,250])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
x=[2,4,6,8]
y=[3,6,9,10]
plt.plot(x,y)
plt.title("LINE PLOT")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x=[2,4,6,8]
y=[3,6,9,10]
plt.plot(x,y)
plt.title("LINE PLOT",fontsize=18,loc='right')
plt.xlabel("x axis",fontsize=18,loc='center')
plt.ylabel("y axis",fontsize=18,loc='top')
plt.show()
```

```python
x=[2,4,6,8]
y=[3,6,9,10]
plt.plot(x,y,color="m",linewidth=3) # u can use the colors first letter also, k for black, b for blue
plt.title("LINE PLOT",fontsize=18,loc='right')
plt.xlabel("x axis",fontsize=18,loc='center')
plt.ylabel("y axis",fontsize=18,loc='top')
plt.show()
```

```python
x=[2,4,6,8]
y=[3,6,9,10]
plt.plot(x,y,color="firebrick",linewidth=4,marker='*',markersize=15,markerfacecolor='red',markeredgecolor='lime',linestyle="-.") 
plt.title("LINE PLOT",fontsize=18,loc='right')
plt.xlabel("x axis",fontsize=18,loc='center')
plt.ylabel("y axis",fontsize=18,loc='center')
plt.show()
```

```python
# another method using axis method by removing the xlim and ylim

import matplotlib.pyplot as plt
import numpy as np
plt.style.use("Solarize_Light2")
x=np.random.randint(0,100,50)
y=np.random.randint(0,50,50)
plt.plot(x,"m:",linewidth=4,markersize=9,markerfacecolor='red',markeredgecolor='line',label='Line1')
plt.plot(y,"c:",linewidth=4,markersize=9,markerfacecolor='red',markeredgecolor='line',label='Line2')
plt.title("LINE PLOT",fontsize=18,loc='center')
plt.xlabel("x axis",fontsize=18,loc='center')
plt.ylabel("y axis",fontsize=18,loc='center')
plt.axis([0,70,0,25])
plt.legend(fontsize=15)
plt.show()
```

```python
plt.style.use("Solarize_Light2")
plt.figure(figsize=(10,10))
x=np.random.randint(0,100,50)
y=np.random.randint(0,50,50)
print(x)
print(y)
plt.plot(x,"m:",linewidth=4,markersize=9,markerfacecolor='red',markeredgecolor='lime',label='Line1')
plt.plot(y,"c:",linewidth=4,markersize=9,markerfacecolor='red',markeredgecolor='lime',label='Line1')
plt.title("LINE PLOT",fontsize=18,loc='center')
plt.xlabel("x axis",fontsize=18,loc='center')
plt.xticks(np.arange(0,50,12))
plt.yticks(np.arange(0,50,2))
plt.legend(fontsize=15)
plt.show()
```
