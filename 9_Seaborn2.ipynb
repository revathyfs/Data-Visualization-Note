Draw the heat mp to show the intensity of corr

```python
data = sns.load_dataset('iris')
sns.heatmap(data.corr(),annot=True)
```

```python
data = sns.load_dataset('iris')

# Select only numeric columns for correlation
numeric_data = data.select_dtypes(include='number')

# Plot the heatmap
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
sns.heatmap(data.corr(),annot=True,linewidth=4,cmap='gnuplot',linecolor='k')
```

```python
# Load dataset
data = sns.load_dataset('iris')

# Select numeric columns only
numeric_data = data.select_dtypes(include='number')

# Plot the heatmap with your desired settings
sns.heatmap(numeric_data.corr(), 
            annot=True, 
            linewidth=4, 
            cmap='gnuplot', 
            linecolor='k')

plt.title("Correlation Heatmap of Iris Numeric Features")
plt.show()
```
