Matplotlib

ðŸ’¡ What is Matplotlib?

Matplotlib is a plotting library for creating static, animated, and interactive visualizations in Python.

```python
# Importing
import matplotlib.pyplot as plt
```

```python
#Plotting

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Other Plot Types

```python
# Bar plot
plt.bar(['A', 'B', 'C'], [3, 7, 1])
plt.show()
```

```python
# Histogram
data = [1, 2, 2, 3, 3, 3, 4]
plt.hist(data, bins=4)
plt.show()
```

```python
# Scatter plot
plt.scatter(x, y)
plt.show()
```

```python
# Pie chart
labels = ['A', 'B', 'C']
sizes = [15, 30, 55]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.show()
```

Seaborn

ðŸ’¡ What is Seaborn?

Seaborn is built on top of Matplotlib and provides a high-level interface for attractive statistical graphics.

```python
# Importing

import seaborn as sns
```

```python
# shows in-build dataset
dataset_names = sns.get_dataset_names()
print(dataset_names)
```

```python
# Dataset & Plots

# Built-in dataset
df = sns.load_dataset('tips')
df.head()
```

```python
# Scatter plot with regression line
sns.lmplot(x="total_bill", y="tip", data=df)
plt.show()
```

```python
# Box plot
sns.boxplot(x="day", y="total_bill", data=df)
plt.show
```

```python
# Count plot
sns.countplot(x="day", data=df)
plt.show()
```

```python
# Histogram / KDE
sns.histplot(df['total_bill'], kde=True)
plt.show
```

```python
# Heatmap (correlation)
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
# NumPy 
#Use the tips dataset to extract numeric array operations:

import numpy as np
import seaborn as sns

tips = sns.load_dataset('tips')
print(tips.head())

# Convert numeric columns to NumPy array
num_cols = tips[['total_bill', 'tip']]
arr = num_cols.to_numpy()

print(arr.shape)  
print("Mean total bill:", np.mean(arr[:,0]))
print("Tip std dev:", np.std(arr[:,1]))

```

Explore:

- Reshaping:â€¯arr.reshape(-1,)

- Broadcasting: adding a constant

- Filtering: arr[arr[:,1] > np.mean(arr[:,1])]

```python
#Pandas
#Use DataFrame power to analyze the dataset:

import pandas as pd

df = tips.copy()
df.info()
df.describe()

# New column: tip percentage
df['tip_pct'] = df['tip'] / df['total_bill'] * 100

# Filter big bills
big_bills = df[df['total_bill'] > 20]
print(big_bills[['total_bill', 'tip_pct']].head())

# Group and aggregate
grouped = df.groupby('day')['tip_pct'].agg(['mean', 'count'])
print(grouped)
```

```python
#Matplotlib
#Plot from scratch with Matplotlib:

import matplotlib.pyplot as plt

# Scatter: total bill vs tip percentage
plt.figure(figsize=(6, 4))
plt.scatter(df['total_bill'], df['tip_pct'], alpha=0.6)
plt.title("Tip % vs Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip (%)")
plt.grid(True)
plt.show()
```

```python
#  Seaborn
# Create polished visuals with Seaborn:

import seaborn as sns

# Load data with tip_pct
df = sns.load_dataset('tips')
df['tip_pct'] = df['tip'] / df['total_bill'] * 100

# lmplot
sns.lmplot(x='total_bill', y='tip_pct', hue='day', data=df, height=5)
plt.title("Tip% vs Bill by Day")

# Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(x='day', y='tip_pct', data=df)
plt.title("Tip % Distribution by Day")

# Heatmap of correlations
corr = df[['total_bill', 'tip', 'tip_pct', 'size']].corr()
plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
